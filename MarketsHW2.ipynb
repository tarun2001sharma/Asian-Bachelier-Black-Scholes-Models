When the drift and volatility are time-dependent, the stochastic differential equation (SDE) of the asset price under the risk-neutral measure is:
$$
dS_t = \mu(t)S_t\,dt + \sigma(t)S_t\,dW_t.
$$

We define:
$$
L_t = \ln(S_t).
$$

**Deriving the SDE for $L_t$.**

Apply It's lemma to $L_t$:
$$
f(S_t) = \ln(S_t), \quad f'(S_t) = \frac{1}{S_t}, \quad f''(S_t) = -\frac{1}{S_t^2}.
$$

Then:
$$
dL_t = f'(S_t)dS_t + \frac{1}{2}f''(S_t)(dS_t)^2.
$$

Substitute $dS_t = \mu(t)S_t\,dt + \sigma(t)S_t\,dW_t$:
$$
dL_t = \frac{1}{S_t}(\mu(t)S_t\,dt + \sigma(t)S_t\,dW_t) + \frac{1}{2}\left(-\frac{1}{S_t^2}\right)(\sigma(t)^2 S_t^2\,dt).
$$

This simplifies to:
$$
dL_t = \mu(t)\,dt + \sigma(t)\,dW_t - \frac{\sigma(t)^2}{2}dt.
$$

Thus:
$$
dL_t = \left(\mu(t) - \frac{\sigma(t)^2}{2}\right)dt + \sigma(t)dW_t.
$$

**Mean and Variance of $L_T$.**

Integrate from $0$ to $T$:
$$
L_T = L_0 + \int_0^T \left(\mu(s) - \frac{\sigma(s)^2}{2}\right)ds + \int_0^T \sigma(s)\,dW_s.
$$

Since $L_0 = \ln(S_0)$ and $W_s$ is a Brownian motion:
- Mean:
$$
E[L_T] = \ln(S_0) + \int_0^T \left(\mu(s) - \frac{\sigma(s)^2}{2}\right)ds.
$$

- Variance comes from the stochastic integral:
$$
\text{Var}(L_T) = \int_0^T \sigma(s)^2 ds.
$$

Hence:
$$
L_T \sim N\left( \ln(S_0) + \int_0^T \left(\mu(s)-\tfrac{\sigma(s)^2}{2}\right)ds,\; \int_0^T \sigma(s)^2 ds \right).
$$

**Forward Price, Undiscounted and Discounted Payoffs.**

1. **Forward Price:**
   Under risk-neutral dynamics, typically $\mu(t) = r(t) - q(t)$, where $r(t)$ is the instantaneous risk-free rate and $q(t)$ is the continuous dividend yield (if any). The forward price is:
   $$
   E_0[S_T] = S_0 \exp\left(\int_0^T \mu(s) ds\right).
   $$
   
   Substituting $\mu(s) = r(s)-q(s)$:
   $$
   E_0[S_T] = S_0 \exp\left(\int_0^T (r(s)-q(s))ds\right).
   $$

2. **Undiscounted Call Payoff $E_0[(S_T - K)^+]$:**
   Since $S_T$ is lognormal with time-dependent parameters, we have:
   - Mean of $L_T$: $m_T = \ln(S_0) + \int_0^T(\mu(s)-\tfrac{\sigma(s)^2}{2})ds$,
   - Variance of $L_T$: $v_T = \int_0^T \sigma(s)^2 ds$.

   Then:
   $$
   (S_T - K)^+ = S_T \mathbf{1}_{\{S_T > K\}}.
   $$
   
   By the standard lognormal option pricing formula (now with integrated parameters),
   $$
   E_0[(S_T - K)^+] = S_0 \exp\left(-\int_0^T q(s)ds\right)\Phi(d_1) - K\Phi(d_2),
   $$
   where
   $$
   d_1 = \frac{m_T - \ln(K)}{\sqrt{v_T}} + \frac{\sqrt{v_T}}{2}, \quad d_2 = d_1 - \sqrt{v_T}.
   $$

3. **Discounted Call Payoff $E_0[e^{-\int_0^T r(s)ds}(S_T - K)^+]$:**
   Applying the discount factor:
   $$
   E_0[e^{-\int_0^T r(s)ds}(S_T - K)^+] = S_0 \exp\left(-\int_0^T q(s)ds\right)\Phi(d_1) - K\exp\left(-\int_0^T r(s)ds\right)\Phi(d_2).
   $$

**Equivalent Constant Parameters.**

To recover a standard (constant-parameter) Black-Scholes model that leads to the same distribution and results, we define effective constants:
- Let:
  $$
  \bar{r} = \frac{1}{T}\int_0^T r(s)ds, \quad \bar{q} = \frac{1}{T}\int_0^T q(s)ds, \quad \bar{\sigma}^2 = \frac{1}{T}\int_0^T \sigma(s)^2 ds.
  $$

- Then $\bar{\mu} = \bar{r} - \bar{q}$.

With these constants, the lognormal distribution at $T$ under a constant-parameter Black-Scholes model:
$$
d\tilde{S}_t = \bar{\mu}\tilde{S}_t\,dt + \bar{\sigma}\tilde{S}_t\,dW_t,
$$
will match the final distribution of $S_T$ from the time-dependent model. This ensures that:
- $\tilde{L}_T \sim N\left(\ln(S_0) + (\bar{\mu} - \tfrac{\bar{\sigma}^2}{2})T,\; \bar{\sigma}^2 T\right)$
matches the original $L_T$ distribution. The option prices computed from this equivalent constant-parameter model will be the same as those from the time-dependent parameters model.

X_t = S_t^N.

f'(S_t) = N S_t^{N-1}, \quad f''(S_t) = N(N-1)S_t^{N-2}.

$$
dX_t = f'(S_t)dS_t + \tfrac{1}{2}f''(S_t)(dS_t)^2.

$$
dX_t = N S_t^{N-1}(\mu S_t\,dt + \sigma S_t\,dW_t) + \tfrac{1}{2}N(N-1)S_t^{N-2}(\sigma^2 S_t^2\,dt).

Simplify powers of $S_t$:
$$
dX_t = N S_t^N \mu\,dt + N S_t^N \sigma\,dW_t + \tfrac{1}{2}N(N-1)S_t^N \sigma^2\,dt.
$$

Factor out $X_t = S_t^N$:
$$
dX_t = X_t \left[ \left( N\mu + \tfrac{N(N-1)}{2}\sigma^2 \right)dt + N\sigma\,dW_t \right].
$$

Thus, the SDE for $X_t = S_t^N$ is:
$$
dX_t = X_t \bigl( \tilde{\mu}\,dt + \tilde{\sigma}\,dW_t \bigr),
$$
where:
$$
\tilde{\mu} = N\mu + \frac{N(N-1)}{2}\sigma^2 \quad\text{and}\quad \tilde{\sigma} = N\sigma.
$$

Under the risk-neutral measure (for simplicity assume no dividends, so $\mu = r$):

1. **Forward Price of $X$:**
   The forward price $F_0(T)$ of $X_T$ is $E_0[X_T]$. Since $X_T$ is lognormal with:
   $$
   X_T = S_0^N \exp\left(N(r - \tfrac{\sigma^2}{2})T + N\sigma W_T\right),

   - Mean of $\ln(X_T)$: $N \ln(S_0) + N(r - \tfrac{\sigma^2}{2})T$

   Therefore:
   $$
   E[X_T] = \exp\left( N\ln(S_0) + N(r - \tfrac{\sigma^2}{2})T + \tfrac{1}{2}N^2 \sigma^2 T \right).
   $$
   Simplify:
   $$
   E[X_T] = S_0^N \exp\left( N r T + \frac{N(N-1)}{2}\sigma^2 T \right).

   The forward price for delivery at $T$ is this expectation with no discounting.

2. **Discounted Expectation of $X_T$:**
   For the discounted expectation $E_0[e^{-rT}X_T]$:
   $$
   E_0[e^{-rT}X_T] = S_0^N \exp\left(N r T + \frac{N(N-1)}{2}\sigma^2 T - rT\right)

3. **Pricing Calls and Puts on $X_T$:**

   $$
   X_T = X_0 \exp\left((r - \tfrac{\tilde{\sigma}^2}{2})T + \tilde{\sigma}W_T\right),
   $$
   where $\tilde{\sigma}^2 = (N\sigma)^2 = N^2\sigma^2$.

   Thus, a European call on $X$ has value at time 0:
   $$
   C_0 = e^{-rT}\left[ X_0 \Phi(d_1) - K \Phi(d_2) \right],
   $$
   where
   $$
   d_1 = \frac{\ln(X_0/K) + (r + \tfrac{\tilde{\sigma}^2}{2})T}{\tilde{\sigma}\sqrt{T}}, \quad
   d_2 = d_1 - \tilde{\sigma}\sqrt{T}.
   $$

   Substitute $X_0 = S_0^N$ and $\tilde{\sigma} = N\sigma$:
   $$
   d_1 = \frac{\ln(S_0^N/K) + \left(r + \frac{N^2\sigma^2}{2}\right)T}{N\sigma\sqrt{T}}
   = \frac{N\ln(S_0) - \ln(K) + \left(r + \frac{N^2\sigma^2}{2}\right)T}{N\sigma\sqrt{T}}.
   $$

   Similarly, a put can be priced using the put-call parity or directly with the Black-Scholes formula for puts.

**Given setup:**
We consider the Bachelier model with zero interest rates:
$$
dS_t = \sigma \, dW_t.
$$
Thus:
$$
S_t = S_0 + \sigma W_t.
$$

**(a)** **Expression for $S_T$:**

Since $W_t \sim N(0,t)$, we have:
$$
S_T = S_0 + \sigma W_T.
$$

Here, $S_T$ is normally distributed with mean $S_0$ and variance $\sigma^2 T$.

---

**(b)** **Price of a vanilla option on $S_T$ with strike $K$:**

Without discounting (as rates are zero), the option price at time 0 is:
$$
\text{Price} = E[(S_T - K)^+] = \int_{K}^{\infty} (x - K) \frac{1}{\sqrt{2\pi\sigma^2 T}}\exp\left(-\frac{(x-S_0)^2}{2\sigma^2 T}\right) dx.
$$

This is the Bachelier formula for a call:
$$
\text{Price} = (S_0 - K)\Phi(d) + \sigma\sqrt{T}\,\phi(d),
$$
where
$$
d = \frac{S_0 - K}{\sigma\sqrt{T}}, \quad \Phi(\cdot) \text{ is the standard normal CDF,} \quad \phi(\cdot) \text{ is the standard normal PDF.}
$$

Since interest rates are zero, discounted and undiscounted prices coincide.

---

**(c)** **Expression for $A^T = \int_0^T S_t \, dt$:**

We know $S_t = S_0 + \sigma W_t$. So:
$$
A^T = \int_0^T S_t \, dt = \int_0^T (S_0 + \sigma W_s) \, ds.
$$

Integrating term by term:
$$
A^T = S_0 T + \sigma \int_0^T W_s \, ds.
$$

---

**(d)** **Rewrite the integral in terms of $dW_s$:**

We use the known representation:
$$
\int_0^T W_s \, ds = \int_0^T (T-u) \, dW_u.
$$

This can be derived by integration by parts or by recognizing that $W_s$ is a Brownian motion. Thus:
$$
A^T = S_0 T + \sigma \int_0^T (T-u) \, dW_u.
$$

---

**(e)** **Writing $A^T$ as the final value of an SDE:**

Define a process $A_t$ by:
$$
A_t = \int_0^t S_u \, du.
$$
Its differential is:
$$
dA_t = S_t \, dt = (S_0 + \sigma W_t) \, dt.
$$

Alternatively, from the $(dW)$ representation:
$$
A_t = S_0 t + \sigma \int_0^t (t-u) dW_u.
$$

If we fix $T$ and consider $A^T$ as a final value at time $T$, we have:
$$
A^T = A_0^T + \int_0^T (S_0 + \sigma W_u)\,du.
$$

For an SDE form with respect to $dW_u$, consider:
$$
d\tilde{A}_t = (T-t) dW_t, \quad \tilde{A}_0=0.
$$

Then:
$$
\tilde{A}_T = \int_0^T (T-u)dW_u.
$$

Hence:
$$
A^T = S_0 T + \sigma \tilde{A}_T.
$$

Here, $\tilde{A}_t$ follows an SDE with "volatility" $(T-t)$ that decreases as $t$ approaches $T$. Comparing the "volatility" structure, $A_t$ has a time-dependent volatility pattern different from the constant volatility in $S_t$.

---

**(f)** **Is $A^T$ normal? Mean and variance of $A^T$:**

$A^T$ is a linear functional of a Gaussian process ($W_t$), hence $A^T$ is normal.

- Mean:
  $$
  E[A^T] = \int_0^T E[S_t] dt = \int_0^T S_0 \, dt = S_0 T.
  $$

- Variance:
  $$
  \text{Var}(A^T) = \sigma^2 \text{Var}\left(\int_0^T W_s ds\right).
  $$

Since $E[W_s] = 0$ and $\text{Cov}(W_s, W_u) = \min(s,u)$:
$$
\text{Var}\left(\int_0^T W_s ds\right) = \int_0^T\int_0^T E[W_s W_u] ds du = \int_0^T\int_0^T \min(s,u) ds du = \frac{T^3}{3}.
$$

Thus:
$$
\text{Var}(A^T) = \sigma^2 \frac{T^3}{3}.
$$

So:
$$
A^T \sim N\bigl(S_0 T,\; \frac{\sigma^2 T^3}{3}\bigr).
$$

---

**(g)** **Mean and variance of $A^T/T$:**

Divide $A^T$ by $T$:
$$
\frac{A^T}{T} = S_0 + \sigma \frac{1}{T}\int_0^T W_s ds.
$$

Since $A^T/T$ is also normal:
- Mean:
  $$
  E\left[\frac{A^T}{T}\right] = S_0.
  $$

- Variance:
  $$
  \text{Var}\left(\frac{A^T}{T}\right) = \frac{\sigma^2 T^3}{3T^2} = \frac{\sigma^2 T}{3}.
  $$

Compare to $S_T$:
- $S_T \sim N(S_0, \sigma^2 T)$.
- $A^T/T \sim N(S_0, \sigma^2 T/3)$.

So $A^T/T$ has the same mean but lower variance than $S_T$.

**Price of a vanilla option on $A^T/T$ with strike $K$:**
With no discounting and $A^T/T \sim N(S_0, \sigma^2 T/3)$, a call option $(A^T/T - K)^+$ is priced similarly to the Bachelier formula but with reduced volatility $\sigma\sqrt{T/3}$. This results in a lower option price compared to the option on $S_T$ because the underlying $A^T/T$ has lower variance than $S_T$.

In summary:
- $A^T/T$ has less volatility than $S_T$.
- A call on $A^T/T$ is cheaper than a call on $S_T$ with the same strike $K$, all else equal.

We consider an Asian contract based on discrete averaging:
$$
A_N = \frac{1}{N}\sum_{i=1}^N S_{T_i}.
$$
We assume zero interest rates ($r=0$) and initially zero dividend yield ($q=0$) for simplicity.

---
**(a)**

**Constructing a trading strategy at zero rates and zero dividends:**

Since $r=0$, the stock price $S_t$ is a martingale under the risk-neutral measure (no drift). The expected value of $S_{T_i}$ is $S_0$. Thus $E[A_N] = S_0$.

We want a self-financing, zero-cost strategy that yields $A_N$ at time $T = T_N$. Consider entering into $N$ forward contracts at time 0, each maturing at $T_i$, on one share of the stock $S$. At $r=0$ and $q=0$, the forward price for delivery at $T_i$ is simply $S_0$.

- Each forward contract has zero initial cost.
- At $T_i$, the forward contract delivers $S_{T_i}$ in exchange for paying $S_0$.

Now, if at $T_i$ we receive $S_{T_i}$ and immediately store it as cash (since $r=0$, holding cash costs nothing and yields no interest), then at time $T_N$ we still have $S_{T_i}$ in cash form.

By doing this for all $i$ from 1 to $N$, at time $T_N$ we have:
$$
\sum_{i=1}^N S_{T_i}.
$$

To get exactly $A_N$, we scale down the position. Instead of entering into $N$ forward contracts each for 1 share, we enter into just one forward contract that delivers the entire sum $\sum_{i=1}^N S_{T_i}$. Alternatively, enter into $N$ forward contracts each for $\frac{1}{N}$ shares of stock.

Hence, at $T_N$:
$$
\frac{1}{N}\sum_{i=1}^N S_{T_i} = A_N.
$$

**Initial quantity of stock and cash:**
A forward contract can be replicated by holding one share of stock and short-selling $S_0$ in cash at time 0. Since $r=0$, $S_0$ dollars at time 0 and $S_0$ dollars at time $T_i$ have the same value. Thus, to create $\frac{1}{N}$ forwards for each $T_i$, we:
- Buy $\frac{1}{N}$ share of stock.
- Short-sell $S_0/N$ units of cash (equivalent to a zero-coupon bond with face value $S_0/N$).

Summing over all $T_i$, we hold:
$$
\sum_{i=1}^N \frac{1}{N}\text{(forward)} = \frac{1}{N} \sum_{i=1}^N (1 \text{ forward at } T_i).
$$

But each forward is separate, so to start with zero cash net, we balance the initial short cash and long stock. The net initial cost of setting up all forwards is zero.

In essence, the initial strategy involves no net initial investment (zero cash), but a combination of long stock and short cash that replicates a series of forwards. At each $T_i$, you "close" the forward, extracting $S_{T_i}$, and hold it as cash until $T_N$. At $T_N$, sum them up and divide by $N$ to get $A_N$.

---

**(b)**

**Interpretation in terms of pricing:**

The above strategy shows that the value of $A_N$ at $T_N$ can be constructed from zero initial cost. Hence, the "price" of a contract paying $A_N$ at time $T_N$ (under zero rates and no dividends) is simply the current stock price $S_0$. This is because the expected value under the risk-neutral measure is $S_0$, and with $r=0$, the discounted value is also $S_0$.

For a forward contract with payoff $A_N - K$ at $T_N$, setting its initial price to zero implies $K = S_0$ because the expected payoff is $A_N - S_0$. If you choose $K=S_0$, the initial cost is zero, which aligns with the no-arbitrage condition.

---

**(c)**

**Nonzero $r$:**

If now $r>0$, the forward price of $S_{T_i}$ is no longer $S_0$, but $S_0 e^{rT_i}$. To replicate $A_N$ at $T_N$ using a similar strategy, you'd enter into forward contracts that cost zero initially but now you need to account for discounting:

- The expected discounted payoff $e^{-rT_N} A_N$ at time 0 is $S_0$, so $A_N$ at $T_N$ has a present value $S_0 e^{rT_N}$.
- To replicate this, you'd hold fewer shares initially (because shares grow on average at the risk-free rate under the pricing measure) and finance the rest with borrowing or lending at rate $r$.

In other words, with $r>0$, the trading strategy must be adjusted so that the final portfolio still yields $A_N$. This reintroduces interest rate terms into the amount of cash or bonds held. The forward price for $A_N$ (deliver $A_N$ at $T_N$) is $S_0 e^{rT_N}$. Similarly, for the contract $A_N - K$, the forward price is zero if $K = S_0 e^{rT_N}$.

---

**(d)**

**Nonzero dividend yield $q$:**

If there is a continuous dividend yield $q>0$, the stock price under the risk-neutral measure has drift $(r-q)$. The forward price of $S_{T_i}$ becomes $S_0 e^{(r-q)T_i}$. To replicate $A_N$, you must adjust the trading strategy to be long more shares initially to offset the dividend yield you lose by not holding the stock directly. Essentially, the presence of $q$ reduces the forward price, and to replicate $A_N$, the quantity of stock held at inception (and at each date) will increase to compensate for the dividend discount.

---

**(e)**

**Determination of the trading strategy:**

Given $r, q, T, N$, we can write down the exact portfolio weights $a_i(t)$ that replicate $A_N$. Each $a_i(t)$ corresponds to holding $\frac{1}{N}$ of a forward-type position on $S_{T_i}$. By adjusting the number of shares held at each sub-interval $[T_{i-1}, T_i]$, you replicate the payoff at maturity.

If you plot $c(t) = \sum_i a_i(t)S_{T_i}$ (the value of the cash position at time $t$) and $a_i(t)$, you'll see how the position in the underlying evolves over time. With nonzero $r$ or $q$, these coefficients $a_i(t)$ differ from the zero-rate, zero-dividend case to ensure the final payoff matches $A_N$ exactly.

---

**(f)**

**Extension to arbitrary weightings:**

If we define a weighted average:
$$
A_N^w = \sum_i w_i S_{T_i}, \quad \text{or in continuous time } A^\Psi = \int_0^T w(s) S_s ds,
$$
where $w_i$ or $w(s)$ are some given weights (that sum or integrate to 1), we can extend the same logic. By constructing a portfolio of forwards or dynamic trading strategies that track $S_t$ at each weighted observation time, we replicate any weighted average payoff. The main difference is just the weighting applied to each $S_{T_i}$ (or continuous integral), and thus the portfolio strategy will hold a time-dependent quantity of shares corresponding to the weighting function $w(s)$.

This demonstrates that any linear functional of the stock price path can be replicated by an appropriate dynamic strategy, given no transaction costs and frictionless markets.
